```python
# 📦 Imports
import pandas as pd
import glob
import os

# 📁 Caminho da pasta com os CSVs
caminho_pasta = '/home/flaviana/Alexandre/compre_up/export_csv/projeto_01/dados'
caminho_csvs = os.path.join(caminho_pasta, '*.csv')

# 📄 Lê todos os CSVs e cria variáveis com base nos nomes dos arquivos
dataframes = {} # Vamos armazenar os DataFrames em um dicionário

for caminho in glob.glob(caminho_csvs):
    nome_arquivo = os.path.basename(caminho).replace('.csv', '')
    try:
        df = pd.read_csv(caminho)
        dataframes[nome_arquivo] = df
        # print(f"✅ Carregado: {nome_arquivo} ({caminho})")
    except Exception as e:
        print(f"❌ Erro ao carregar {nome_arquivo} ({caminho}): {e}")

# ✅ Agora vamos exibir o conteúdo de cada DataFrame carregado
print("\n--- Conteúdo dos DataFrames Carregados ---")
for nome, df in dataframes.items():
    print(f"\nDataFrame: {nome}")
    display(df) # Use print(df) se não estiver em um ambiente interativo
```

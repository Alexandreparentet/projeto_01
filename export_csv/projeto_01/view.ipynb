```python
# üì¶ Imports
import pandas as pd
import glob
import os

# üìÅ Caminho da pasta com os CSVs
caminho_pasta = '/home/flaviana/Alexandre/compre_up/export_csv/projeto_01/dados'
caminho_csvs = os.path.join(caminho_pasta, '*.csv')

# üìÑ L√™ todos os CSVs e cria vari√°veis com base nos nomes dos arquivos
dataframes = {} # Vamos armazenar os DataFrames em um dicion√°rio

for caminho in glob.glob(caminho_csvs):
    nome_arquivo = os.path.basename(caminho).replace('.csv', '')
    try:
        df = pd.read_csv(caminho)
        dataframes[nome_arquivo] = df
        # print(f"‚úÖ Carregado: {nome_arquivo} ({caminho})")
    except Exception as e:
        print(f"‚ùå Erro ao carregar {nome_arquivo} ({caminho}): {e}")

# ‚úÖ Agora vamos exibir o conte√∫do de cada DataFrame carregado
print("\n--- Conte√∫do dos DataFrames Carregados ---")
for nome, df in dataframes.items():
    print(f"\nDataFrame: {nome}")
    display(df) # Use print(df) se n√£o estiver em um ambiente interativo
```
